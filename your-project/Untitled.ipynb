```python


gmaps.configure(api_key)

```

```python
# Example locations and air quality data
locations = [(37.7749, -122.4194), (34.0522, -118.2437), (40.7128, -74.0060)]  # Example coordinates (San Francisco, Los Angeles, New York)
air_quality_data = [20, 30, 40]  # Example air quality data for each location

fig = gmaps.figure()
heatmap_layer = gmaps.heatmap_layer(locations, weights=air_quality_data, max_intensity=50, point_radius=10)
fig.add_layer(heatmap_layer)
display(fig)
```

```python
import folium
from folium.plugins import HeatMap
from IPython.display import display

# Create a map centered at San Francisco
m = folium.Map(location=[37.7749, -122.4194], zoom_start=10)

# Example allergen sources
allergen_sources = [
    {"name": "Tree Pollen", "location": [37.7805, -122.4233]},
    {"name": "Grass Pollen", "location": [37.7739, -122.4313]},
    {"name": "Ragweed Pollen", "location": [37.7800, -122.4337]}
]

# Add markers for allergen sources
for source in allergen_sources:
    folium.Marker(location=source["location"], popup=source["name"]).add_to(m)

# Example heatmap data
heatmap_data = [
    [37.7805, -122.4233, 10],  # [lat, lon, intensity]
    [37.7739, -122.4313, 20],
    [37.7800, -122.4337, 30]
]

# Add HeatMap layer to the map
HeatMap(heatmap_data).add_to(m)

# Display the map
display(m)
```

```python
import requests

url = 'https://pollen.googleapis.com/v1/forecast:lookup'


params = {
    'key': api_key,
    'location.longitude': 35.32,
    'location.latitude': 32.32,
    'days': 1
}

response = requests.get(url, params=params)

if response.ok:
    pollen_data = response.json()
    print(pollen_data)
else:
    print(f"Error: {response.status_code} - {response.text}")
```
